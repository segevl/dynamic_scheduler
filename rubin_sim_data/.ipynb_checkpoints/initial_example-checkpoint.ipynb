```python
# Set the survey start time. Presumably we will change this to when we have cloud data
SURVEY_START_MJD = Time("2026-02-24T12:00:00").mjd
nside = 32
```

```python
# Quick copy-paste so we can generate our current standard survey footprint.
def standard_survey_footprint(survey_start_mjd=SURVEY_START_MJD, nside=32, nslice=2,
                              rolling_scale=0.9, rolling_uniform=True):
    # Generate footprint over the sky
    sky = CurrentAreaMap(nside=nside)
    footprints_hp_array, labels = sky.return_maps()
    # Identify pixels for rolling
    roll_indx = np.where((labels == "lowdust") | (labels == "virgo"))[0]
    roll_footprint = footprints_hp_array["r"] * 0
    roll_footprint[roll_indx] = 1

    footprints_hp = {}
    for key in footprints_hp_array.dtype.names:
        footprints_hp[key] = footprints_hp_array[key]

    # Set up a mask to contain some surveys within this region
    footprint_mask = footprints_hp["r"] * 0
    footprint_mask[np.where(footprints_hp["r"] > 0)] = 1

    # Use the Almanac to find the position of the sun at the start of survey
    almanac = Almanac(mjd_start=survey_start_mjd)
    sun_moon_info = almanac.get_sun_moon_positions(survey_start_mjd)
    sun_ra_start = sun_moon_info["sun_RA"].copy()

    # Define the rolling footprint
    footprints = make_rolling_footprints(
        fp_hp=footprints_hp,
        mjd_start=survey_start_mjd,
        sun_ra_start=sun_ra_start,
        nslice=nslice,
        scale=rolling_scale,
        nside=nside,
        wfd_indx=roll_indx,
        order_roll=1,
        n_cycles=3,
        uniform=rolling_uniform,
    )

    return footprints
```

```python
# Taken from https://github.com/lsst-sims/sims_featureScheduler_runs5.2/blob/main/comp_survey/lsst_surveys.py

# Set up values to use as kwarg defaults.
NEXP = 1
U_NEXP = 1
EXPTIME = 30.0
U_EXPTIME = 38.0
GOOD_SEEING_MAX = 1.3
CAMERA_ROT_LIMITS = (-80.0, 80.0)
SCIENCE_PROGRAM = "BLOCK-407"

def safety_masks(
    nside: int = DEFAULT_NSIDE,
    moon_distance: float = 30,
    wind_speed_maximum: float = 20.0,
    min_alt: float = 20,
    max_alt: float = 86.5,
    min_az: float = 0,
    max_az: float = 360,
    shadow_minutes: float = 70,
    min_az_sunrise: float = 120,
    max_az_sunrise: float = 290,
    time_to_sunrise: float = 3.0,
    apply_time_limited_shadow: bool = True,
    sun_alt_limit: float | None = None,
) -> list[bf.BaseBasisFunction]:
    """Basic safety mask basis functions.

    Avoids the moon, bright planets, high wind, and
    areas on the sky out of bounds, using
    the MoonAvoidanceBasisFunction, PlanetMaskBasisFunction,
    AvoidDirectWindBasisFunction, and the AltAzShadowMaskBasisFunction.
    Adds the default AltAzShadowMaskTimeLimited basis function to avoid
    pointing toward sunrise late in the night during commissioning.

    Parameters
    ----------
    nside : `int` or None
        The healpix nside to use.
        Default of None uses rubin_scheduler.utils.get_default_nside.
    moon_distance : `float`, optional
        Moon avoidance distance, in degrees.
    wind_speed_maximum : `float`, optional
        Wind speed maximum to apply to the wind avoidance basis function,
        in m/s.
    min_alt : `float`, optional
        Minimum altitude (in degrees) to observe.
    max_alt : `float`, optional
        Maximum altitude (in degrees) to observe.
    min_az : `float`, optional
        Minimum azimuth angle (in degrees) to observe.
    max_az : `float`, optional
        Maximum azimuth angle (in degrees) to observe.
    shadow_minutes : `float`, optional
        Avoid inaccessible alt/az regions, as well as parts of the sky
        which will move into those regions within `shadow_minutes` (minutes).
        Should be set to the expected time needed to execute the
        observing block associated with the survey.
    min_az_sunrise : `float`, optional
        Minimum azimuth angle (in degrees) to observe during time period
        at the end of the night (during time_to_sunrise).
    max_az_sunrise: `float`, optional
        Maximum azimuth angle (in degrees) to observe during the time period
        at the end of the night (during time_to_sunrise).
    time_to_sunrise : `float`, optional
        Hours before daybreak (sun @ alt=0) to start the azimuth avoidance
        mask.
    apply_time_limited_shadow : `bool`, optional
        Flag for whether to apply the morning (time_to_sunrise) azimuth mask.
    sun_alt_limit : `float` or None, optional
        Maximum sun altitude (deg) required before proposing targets.

    Returns
    -------
    mask_basis_functions : `list` [`BaseBasisFunction`]
        Mask basis functions should always be used with a weight of 0.
        The masked (np.nan or -np.inf) regions will remain masked,
        but the basis function values won't influence the reward.
    """
    mask_bfs = []
    # Avoid the moon - too close to the moon will trip the REBs
    mask_bfs.append(
        bf.MoonAvoidanceBasisFunction(nside=nside, moon_distance=moon_distance)
    )
    # Avoid bright planets
    mask_bfs.append(bf.PlanetMaskBasisFunction(nside=nside))
    # Avoid the wind
    mask_bfs.append(
        bf.AvoidDirectWind(nside=nside, wind_speed_maximum=wind_speed_maximum)
    )
    # Avoid the alt/az limits - this will pick up limits from the
    # yaml file configurations for the summit as well
    mask_bfs.append(
        bf.AltAzShadowMaskBasisFunction(
            nside=nside,
            min_alt=min_alt,
            max_alt=max_alt,
            min_az=min_az,
            max_az=max_az,
            shadow_minutes=shadow_minutes,
        )
    )

    if apply_time_limited_shadow:
        # Only look away from the azimuth of the sun in the next day
        # permitting emergency dome closure
        mask_bfs.append(
            bf.AltAzShadowTimeLimitedBasisFunction(
                nside=nside,
                min_alt=min_alt,
                max_alt=max_alt,
                min_az=min_az_sunrise,
                max_az=max_az_sunrise,
                shadow_minutes=shadow_minutes,
                # Time until/after sun_keys in hours
                time_to_sun=time_to_sunrise + shadow_minutes / 60.0,
                # 'sunrise' is 0 degree sunrise
                sun_keys=["sunrise"],
            )
        )

    if sun_alt_limit is not None:
        mask_bfs.append(bf.SunAltLimitBasisFunction(alt_limit=sun_alt_limit))

    return mask_bfs


def standard_bf(
    nside: int = DEFAULT_NSIDE,
    bandname: str = "g",
    bandname2: str | None = "i",
    m5_weight: float = 6.0,
    footprint_weight: float = 1.5,
    slewtime_weight: float = 3.0,
    stayband_weight: float = 3.0,
    footprints: Footprints | None = None,
    fiducial_fwhm: float = 1.3,
    season: float = 365.25,
    season_start_hour: float = -4.0,
    season_end_hour: float = 2.0,
    strict: bool = True,
    seeing_fwhm_max: float | None = None,
) -> list[tuple[bf.BaseBasisFunction, float]]:
    """Generate the standard basis functions that are shared by blob surveys

    Parameters
    ----------
    nside : `int`
        The HEALpix nside to use. Defaults to DEFAULT_NSIDE
    bandname : `str`
        The band name for the first observation. Default "g".
    bandname2 : `str`
        The band name for the second in the pair (None if unpaired).
        Default "i".
    m5_weight : `float`
        The weight for the 5-sigma depth difference basis function.
    footprint_weight : `float`
        The weight on the survey footprint basis function.
    slewtime_weight : `float`
        The weight on the slewtime basis function.
    stayband_weight : `float`
        The weight on basis function that tries to stay avoid band changes.
    footprints : `rubin_scheduler.scheduler.utils.Footprints` object
        The desired footprints object. Default of None will work, but is likely
        not desirable.
    fiducial_fwhm : `float`
        The fiducial FWHM for the M5Diff Basis function.
    n_obs_template : `dict`
        The number of observations to take every season in each band.
    season : `float`
        The length of season (i.e., how long before templates expire) (days).
        Default 365.25.
    season_start_hour : `float`
        Hour angle limits to use when gathering templates.
        Default -4 (hours)
    season_end_hour : `float`
       Hour angle limits to use when gathering templates.
       Default +2 (hours)
    strict : `bool`
        If False, use BandChangeBasisFunction which rewards visits in the
        same bandpass as currently in-use.
        If True, use a StrictBandBasisFunction which rewards visits in the
        same bandpass as currently in-use, but also rewards visits in
        different filters if the moon rose/set or twilight ended/started,
        or if there was a large gap in observing.
    seeing_fwhm_max : `float`
        Seeing limit to pass to the FootprintBasisFunction - visits
        with delivered image quality > seeing_fwhm_max will not be counted.

    Returns
    -------
    basis_functions_weights : `list`
        list of tuple pairs (basis function, weight) that is
        (rubin_scheduler.scheduler.BasisFunction object, float)

    """

    bfs = []

    if bandname2 is not None:
        bfs.append(
            (
                bf.M5DiffBasisFunction(
                    bandname=bandname, nside=nside, fiducial_FWHMEff=fiducial_fwhm
                ),
                m5_weight / 2.0,
            )
        )
        bfs.append(
            (
                bf.M5DiffBasisFunction(
                    bandname=bandname2, nside=nside, fiducial_FWHMEff=fiducial_fwhm
                ),
                m5_weight / 2.0,
            )
        )

    else:
        bfs.append(
            (
                bf.M5DiffBasisFunction(
                    bandname=bandname, nside=nside, fiducial_FWHMEff=fiducial_fwhm
                ),
                m5_weight,
            )
        )

    if bandname2 is not None:
        bfs.append(
            (
                bf.FootprintBasisFunction(
                    bandname=bandname,
                    footprint=footprints,
                    out_of_bounds_val=np.nan,
                    nside=nside,
                    seeing_fwhm_max=seeing_fwhm_max,
                ),
                footprint_weight / 2.0,
            )
        )
        bfs.append(
            (
                bf.FootprintBasisFunction(
                    bandname=bandname2,
                    footprint=footprints,
                    out_of_bounds_val=np.nan,
                    nside=nside,
                    seeing_fwhm_max=seeing_fwhm_max,
                ),
                footprint_weight / 2.0,
            )
        )
    else:
        bfs.append(
            (
                bf.FootprintBasisFunction(
                    bandname=bandname,
                    footprint=footprints,
                    out_of_bounds_val=np.nan,
                    nside=nside,
                    seeing_fwhm_max=seeing_fwhm_max,
                ),
                footprint_weight,
            )
        )

    bfs.append(
        (
            bf.SlewtimeBasisFunction(bandname=bandname, nside=nside),
            slewtime_weight,
        )
    )
    if strict:
        bfs.append((bf.StrictBandBasisFunction(bandname=bandname), stayband_weight))
    else:
        bfs.append((bf.BandChangeBasisFunction(bandname=bandname), stayband_weight))

    bandnames = [fn for fn in [bandname, bandname2] if fn is not None]
    bfs.append((bf.BandLoadedBasisFunction(bandnames=bandnames), 0))

    return bfs



def gen_greedy_surveys(
    nside: int = DEFAULT_NSIDE,
    bands: list[str] = ["r", "i", "z", "y"],
    ignore_obs: list[str] = ["DD", "twilight_near_sun", "ToO"],
    camera_rot_limits: tuple[float, float] = CAMERA_ROT_LIMITS,
    exptime: float = EXPTIME,
    nexp: int = NEXP,
    u_exptime: float = U_EXPTIME,
    u_nexp: int = U_NEXP,
    shadow_minutes: float = 15.0,
    max_alt: float = 76.0,
    m5_weight: float = 3.0,
    footprint_weight: float = 0.75,
    slewtime_weight: float = 3.0,
    stayband_weight: float = 100.0,
    repeat_weight: float = -1.0,
    footprints: Footprints | None = None,
    science_program: str = SCIENCE_PROGRAM,
    safety_mask_params: dict | None = None,
) -> list[GreedySurvey]:
    """Generate greedy (single-best choice visits) Surveys.

    Parameters
    ----------
    nside : `int`
        The HEALpix nside to use
    bands : `list` [ `str` ]
        Bands in which to generate greedy surveys.
        Default ['r', 'i', 'z', 'y'].
    ignore_obs : `str` or `list` of `str`
        Ignore observations by surveys that include the given substring(s).
    camera_rot_limits : `list` [ `float` ]
        The limits to impose when rotationally dithering the camera (degrees).
        Default [-80., 80.].
    exptime : `float`
        The exposure time for grizy visits.
    nexp : `int`
        The number of exposures per visit for grizy visits.
    u_exptime : `float`
        The exposure time for u band visits.
    u_nexp : `int`
        The number of exposures per visit for u band visits.
    shadow_minutes : `float`
        Used to mask regions around zenith (minutes).
    max_alt : `float`
        The maximium altitude to use when masking zenith (degrees).
    m5_weight : `float`
        The weight for the 5-sigma depth difference basis function.
    footprint_weight : `float`
        The weight on the survey footprint basis function.
    slewtime_weight : `float`
        The weight on the slewtime basis function.
    stayband_weight : `float`
        The weight on basis function that tries to stay avoid band changes.
    repeat_weight : `float`
        Weight that enhances (+ values) or decreases (- values) the likelihood
        of revisiting the same pointing within a two-hour time gap.
    footprints : `rubin_scheduler.scheduler.utils.footprints.Footprints`
        The footprints to be used for the long-gaps surveys.
    science_program : `str`
        The science_program to use for visits from these surveys.
    safety_mask_params : `dict` or None
        A dictionary of additional kwargs to mass to the standard safety masks.
    """
    # Define the extra parameters that are used in the greedy survey. I
    # think these are fairly set, so no need to promote to utility func kwargs
    greed_survey_params = {
        "block_size": 1,
        "smoothing_kernel": None,
        "seed": 42,
        "camera": "LSST",
        "dither": "night",
    }
    if safety_mask_params is None:
        safety_mask_params = {"nside": nside}
    else:
        safety_mask_params = copy.deepcopy(safety_mask_params)
    if (
        "shadow_minutes" not in safety_mask_params
        or safety_mask_params["shadow_minutes"] < shadow_minutes
    ):
        safety_mask_params["shadow_minutes"] = shadow_minutes

    surveys = []
    detailer_list = [
        detailers.CameraRotDetailer(
            min_rot=np.min(camera_rot_limits), max_rot=np.max(camera_rot_limits)
        )
    ]
    detailer_list.append(detailers.LabelRegionsAndDDFs())

    if "u" in bands:
        detailer_list.append(
            detailers.BandNexp(bandname="u", nexp=u_nexp, exptime=u_exptime)
        )

    for bandname in bands:
        bfs = []
        bfs.extend(
            standard_bf(
                nside,
                bandname=bandname,
                bandname2=None,
                m5_weight=m5_weight,
                footprint_weight=footprint_weight,
                slewtime_weight=slewtime_weight,
                stayband_weight=stayband_weight,
                footprints=footprints,
                strict=False,
            )
        )

        # XXX-magic numbers
        bfs.append(
            (
                bf.VisitRepeatBasisFunction(
                    gap_min=0, gap_max=2 * 60.0, bandname=None, nside=nside, npairs=20
                ),
                repeat_weight,
            )
        )
        # Masks, give these 0 weight
        bfs.append(
            (
                bf.AltAzShadowMaskBasisFunction(
                    nside=nside, shadow_minutes=shadow_minutes, max_alt=max_alt, pad=3.0
                ),
                0,
            )
        )

        masks = safety_masks(nside=nside, shadow_minutes=shadow_minutes)
        for m in masks:
            bfs.append((m, 0))

        weights = [val[1] for val in bfs]
        basis_functions = [val[0] for val in bfs]
        surveys.append(
            GreedySurvey(
                basis_functions,
                weights,
                exptime=exptime,
                bandname=bandname,
                nside=nside,
                ignore_obs=ignore_obs,
                nexp=nexp,
                detailers=detailer_list,
                survey_name=f"greedy {bandname}",
                science_program=science_program,
                observation_reason=f"singles_{bandname}",
                **greed_survey_params,
            )
        )

    return surveys
```

```python
# Build a single-band greedy scheduler
footprint = standard_survey_footprint()
surveys = gen_greedy_surveys(bands=["r"], footprints=footprint)
```

```python
# Make a cloud_maps object
cloud_maps = CloudMap()
# add a cloud frame
cloud_maps.add_frame(np.zeros(hp.nside2npix(nside)), SURVEY_START_MJD)
```

```python
"""
dream_scheduler_comparison.py
==============================
Compares three pointing strategies using REAL DREAM cloud data:

  1. Greedy scheduler (rubin_scheduler GreedySurvey) — standard survey pointing
  2. DREAM "absolute minimum extinction" — always point to clearest sky patch
  3. DREAM "cloud motion tracking" — predict where clear patch will drift

All cloud data loaded from real DREAM HDF5 frames via the ConsDB/LFA pipeline
(identical to the analysis pipeline in the DREAM doc).

Usage
-----
    python dream_scheduler_comparison.py --csv feb5_data.csv

    # Or from a Jupyter cell:
    from dream_scheduler_comparison import main
    main(csv_file="feb5_data.csv")

Outputs (in ./outputs/)
-----------------------
    comparison_metrics_<NIGHT>.png
    cloud_maps_<NIGHT>.png
    coverage_map_<NIGHT>.png
    summary_table_<NIGHT>.csv
"""

from __future__ import annotations
import argparse
import io
import os
import warnings
import numpy as np
import pandas as pd
import h5py
import healpy as hp
import matplotlib
matplotlib.use("Agg")
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter, shift as ndimage_shift
from astropy.coordinates import SkyCoord, EarthLocation, AltAz
from astropy.time import Time
import astropy.units as u
# lsst.resources is optional — we implement the same fetch logic with stdlib
try:
    from lsst.resources import ResourcePath as _ResourcePath
    def _read_url(url: str) -> bytes:
        rp = _ResourcePath(url)
        with rp.open("rb") as fd:
            return fd.read()
except ModuleNotFoundError:
    import urllib.request
    import urllib.parse
    try:
        import boto3
        import botocore
        _boto3_available = True
    except ModuleNotFoundError:
        _boto3_available = False

    def _read_url(url: str) -> bytes:
        """
        Fetch HDF5 bytes from:
          s3://lfa@<bucket>/<key>   → boto3 (anonymous or env-credential)
          https://...               → urllib
          file:// or /path/...      → local filesystem
        """
        if url.startswith("s3://"):
            if not _boto3_available:
                raise RuntimeError(
                    "boto3 is required to fetch s3:// URLs. "
                    "Install it with:  pip install boto3")
            # s3://lfa@bucket/key  or  s3://bucket/key
            # Strip leading "s3://" then optional "lfa@" profile prefix
            rest = url[5:]
            if "@" in rest.split("/")[0]:
                _profile, rest = rest.split("@", 1)
            bucket, *parts = rest.split("/")
            key = "/".join(parts)
            s3 = boto3.client(
                "s3",
                config=botocore.config.Config(signature_version=botocore.UNSIGNED),
            )
            buf = io.BytesIO()
            s3.download_fileobj(bucket, key, buf)
            return buf.getvalue()

        elif url.startswith("https://") or url.startswith("http://"):
            with urllib.request.urlopen(url, timeout=60) as resp:
                return resp.read()

        else:
            # Local file
            path = url.removeprefix("file://")
            with open(path, "rb") as fh:
                return fh.read()

warnings.filterwarnings("ignore")

# ── rubin_scheduler ──────────────────────────────────────────────────────────
import rubin_scheduler.scheduler.basis_functions as bf
import rubin_scheduler.scheduler.detailers as detailers
from rubin_scheduler.scheduler.surveys import GreedySurvey
from rubin_scheduler.scheduler.utils import (
    CurrentAreaMap,
    make_rolling_footprints,
)
from rubin_scheduler.site_models import Almanac, CloudMap
from rubin_scheduler.scheduler.schedulers import CoreScheduler
from rubin_scheduler.scheduler.model_observatory import ModelObservatory
from rubin_scheduler.scheduler import sim_runner

# ─────────────────────────────────────────────────────────────────────────────
# CONSTANTS  (from DREAM analysis doc)
# ─────────────────────────────────────────────────────────────────────────────
NSIDE             = 32
NEST              = True
RUBIN_LAT         = -30.244639
RUBIN_LON         = -70.749417
RUBIN_HEIGHT_M    = 2663.0

BIN_SIZE_KM       = 1000
R_PROJECTION      = 10_000.0        # km

RUBIN_EFFECTIVE_AREA      = np.pi * (6.4 / 2) ** 2   # m²
RUBIN_QUANTUM_EFFICIENCY  = 0.8
EXPOSURE_TIME             = 30.0    # s
READOUT_TIME              = 2.0     # s
SLOT_TIME                 = EXPOSURE_TIME + READOUT_TIME  # 32 s
PHOTON_FLUX_MAG20_ZENITH  = 100.0   # photons s⁻¹ m⁻² QE-weighted

MAX_SLEW_SPEED_ALT = 3.5   # deg/s
MAX_SLEW_SPEED_AZ  = 7.0   # deg/s
SLEW_SETTLE_TIME   = 2.0   # s

MIN_ALT_DEG    = 15.0
MAX_SIGMA_MAG  = 0.3
MAX_FLAG_VALUE = 0

URL_COL  = "lsst.sal.DREAM.logevent_largeFileObjectAvailable.url"
TIME_COL = "time"

CAMERA_ROT_LIMITS = (-80.0, 80.0)
SCIENCE_PROGRAM   = "BLOCK-407"


# ─────────────────────────────────────────────────────────────────────────────
# URL / HDF5 LOADERS  (verbatim from DREAM doc)
# ─────────────────────────────────────────────────────────────────────────────

def transform_url(url: str) -> str:
    url = str(url).strip()
    if url.startswith("https://s3.cp.lsst.org/"):
        return url.replace("https://s3.cp.lsst.org/", "s3://lfa@")
    return url


def fetch_sys_map(url: str) -> tuple[np.ndarray, np.ndarray]:
    """Load cloud_sys HDF5. Returns (clouds, sigma) with bad pixels → NaN."""
    data = _read_url(url)
    with h5py.File(io.BytesIO(data), "r") as f:
        clouds        = np.array(f["clouds"],        dtype=float).ravel()
        sigma         = np.array(f["sigma"],         dtype=float).ravel()
        flags         = np.array(f["flags"],         dtype=int  ).ravel()
        mask_visible  = np.array(f["mask_visible"],  dtype=bool ).ravel()
        nobs          = np.array(f["nobs"],           dtype=int  ).ravel()
    bad = (
        ~mask_visible
        | (nobs == 0)
        | (flags > MAX_FLAG_VALUE)
        | (sigma > MAX_SIGMA_MAG)
        | ~np.isfinite(clouds)
    )
    clouds[bad] = np.nan
    sigma [bad] = np.nan
    return clouds, sigma


# ─────────────────────────────────────────────────────────────────────────────
# COORDINATE UTILITIES  (from DREAM doc)
# ─────────────────────────────────────────────────────────────────────────────

def _rubin_loc() -> EarthLocation:
    return EarthLocation(lat=RUBIN_LAT*u.deg,
                         lon=RUBIN_LON*u.deg,
                         height=RUBIN_HEIGHT_M*u.m)


def radec_to_altaz(ra_deg: float, dec_deg: float, obstime) -> tuple[float, float]:
    t   = Time(obstime) if not isinstance(obstime, Time) else obstime
    sky = SkyCoord(ra=ra_deg*u.deg, dec=dec_deg*u.deg, frame="icrs")
    aa  = sky.transform_to(AltAz(obstime=t.utc, location=_rubin_loc()))
    return float(aa.alt.deg), float(aa.az.deg % 360.0)


def altaz_to_xy(alt_deg: float, az_deg: float
                ) -> tuple[float | None, float | None]:
    if alt_deg < MIN_ALT_DEG:
        return None, None
    alt_r = np.radians(alt_deg)
    az_r  = np.radians(az_deg)
    scale = R_PROJECTION / np.sin(alt_r)
    return (-np.cos(alt_r) * np.sin(az_r) * scale,
             np.cos(alt_r) * np.cos(az_r) * scale)


def xy_to_altaz(x_km: float, y_km: float) -> tuple[float, float]:
    r   = np.sqrt(x_km**2 + y_km**2)
    alt = float(np.degrees(np.arctan2(R_PROJECTION, r)))
    az  = float(np.degrees(np.arctan2(-x_km, y_km)) % 360.0)
    return alt, az


def healpix_to_altaz_vals(mp: np.ndarray, nside: int, obstime
                           ) -> tuple[np.ndarray, np.ndarray, np.ndarray]:
    npix       = hp.nside2npix(nside)
    pix        = np.arange(npix)
    theta, phi = hp.pix2ang(nside, pix, nest=NEST)
    ra         = np.degrees(phi)
    dec        = 90.0 - np.degrees(theta)
    t          = Time(obstime) if not isinstance(obstime, Time) else obstime
    sky        = SkyCoord(ra=ra*u.deg, dec=dec*u.deg, frame="icrs")
    aa         = sky.transform_to(AltAz(obstime=t.utc, location=_rubin_loc()))
    vals       = np.array(mp, dtype=float)
    vals       = np.where(vals == hp.UNSEEN, np.nan, vals)
    return aa.az.deg % 360.0, aa.alt.deg, vals


def _healpix_to_grid(mp: np.ndarray, obstime, bin_size: int = BIN_SIZE_KM):
    az, alt, vals = healpix_to_altaz_vals(mp, NSIDE, obstime)
    alt_r = np.radians(alt)
    az_r  = np.radians(az)
    above = alt > MIN_ALT_DEG
    scale = np.where(above, R_PROJECTION / np.sin(alt_r), np.nan)
    xf    = -np.cos(alt_r) * np.sin(az_r) * scale
    yf    =  np.cos(alt_r) * np.cos(az_r) * scale
    vf    = np.where(above, vals, np.nan)

    r_all = np.sqrt(xf**2 + yf**2)
    c     = r_all <= 15_000.0
    ok    = ~np.isnan(vf[c])

    x_edges = np.arange(-15_000, 15_001, bin_size)
    y_edges = np.arange(-15_000, 15_001, bin_size)
    Hs, _, _ = np.histogram2d(xf[c][ok], yf[c][ok],
                               bins=[x_edges, y_edges], weights=vf[c][ok])
    Hc, _, _ = np.histogram2d(xf[c][ok], yf[c][ok], bins=[x_edges, y_edges])
    with np.errstate(divide="ignore", invalid="ignore"):
        H = Hs / Hc
    H[Hc == 0] = np.nan
    H = H.T
    xc, yc  = (x_edges[:-1]+x_edges[1:])/2, (y_edges[:-1]+y_edges[1:])/2
    Xg, Yg  = np.meshgrid(xc, yc)
    H[np.sqrt(Xg**2 + Yg**2) > 15_000] = np.nan
    return H, Xg, Yg, x_edges, y_edges


def process_to_grid(clouds: np.ndarray, sigma: np.ndarray, obstime):
    ext_g, Xg, Yg, xe, ye = _healpix_to_grid(clouds, obstime)
    sig_g, *_              = _healpix_to_grid(sigma,  obstime)
    return ext_g, sig_g, Xg, Yg, xe, ye


# ─────────────────────────────────────────────────────────────────────────────
# GRID VALUE HELPERS
# ─────────────────────────────────────────────────────────────────────────────

def get_value_at_position(grid: np.ndarray, x_km: float, y_km: float) -> float:
    xi = int(round((x_km / BIN_SIZE_KM) + 15))
    yi = int(round((y_km / BIN_SIZE_KM) + 15))
    if 0 <= xi < grid.shape[1] and 0 <= yi < grid.shape[0]:
        return float(grid[yi, xi])
    return np.nan


def find_absolute_minimum(grid: np.ndarray) -> tuple[float, float, float]:
    if not np.any(~np.isnan(grid)):
        return 0.0, 0.0, np.nan
    idx    = np.nanargmin(grid)
    yi, xi = np.unravel_index(idx, grid.shape)
    return (xi - 15) * BIN_SIZE_KM, (yi - 15) * BIN_SIZE_KM, float(grid[yi, xi])


# ─────────────────────────────────────────────────────────────────────────────
# CLOUD MOTION TRACKING  (from DREAM doc)
# ─────────────────────────────────────────────────────────────────────────────

def compute_motion_with_correlation(grid1: np.ndarray, grid2: np.ndarray,
                                    sigma: float = 5.0,
                                    search_range: int = 10
                                    ) -> tuple[float, float, float]:
    g1s = gaussian_filter(np.nan_to_num(grid1, nan=0), sigma=sigma)
    g2s = gaussian_filter(np.nan_to_num(grid2, nan=0), sigma=sigma)
    m1  = ~np.isnan(grid1) & (grid1 != 0)
    best_corr, best_dx, best_dy = -np.inf, 0, 0
    for dy in range(-search_range, search_range + 1):
        for dx in range(-search_range, search_range + 1):
            sh  = ndimage_shift(g2s, (dy, dx), order=1, mode="constant", cval=0)
            sm  = ndimage_shift((~np.isnan(grid2)).astype(float),
                                (dy, dx), order=0, mode="constant", cval=0) > 0.5
            ok  = m1 & sm
            if ok.sum() < 100:
                continue
            v1, v2 = g1s[ok], sh[ok]
            if np.std(v1) > 0 and np.std(v2) > 0:
                corr = float(np.corrcoef(v1, v2)[0, 1])
                if corr > best_corr:
                    best_corr, best_dx, best_dy = corr, dx, dy
    return float(best_dx), float(best_dy), float(best_corr)


def predict_future_position(cx: float, cy: float,
                             dx_pix: float, dy_pix: float,
                             current_grid: np.ndarray,
                             fallback_threshold: float = 0.5
                             ) -> tuple[float, float]:
    nx = cx - dx_pix * BIN_SIZE_KM
    ny = cy - dy_pix * BIN_SIZE_KM
    r  = np.sqrt(nx**2 + ny**2)
    if r > 14_000:
        nx *= 14_000 / r
        ny *= 14_000 / r
    val = get_value_at_position(current_grid, nx, ny)
    if np.isnan(val) or val > fallback_threshold:
        nx, ny, _ = find_absolute_minimum(current_grid)
    return nx, ny


def compute_all_positions(all_grids: list[np.ndarray]):
    x0, y0, _ = find_absolute_minimum(all_grids[0])
    abs_pos    = [(x0, y0)]
    pred_pos   = [(x0, y0)]
    motion_v   = [(0, 0)]
    xp, yp     = x0, y0
    HL         = 3

    for i in range(1, len(all_grids)):
        if i % 50 == 0:
            print(f"    Tracking frame {i}/{len(all_grids)}")
        xa, ya, _ = find_absolute_minimum(all_grids[i])
        abs_pos.append((xa, ya))

        if i >= HL:
            dxs, dys = [], []
            for j in range(1, HL + 1):
                dx, dy, conf = compute_motion_with_correlation(
                    all_grids[i-j], all_grids[i-j+1])
                if conf > 0.5:
                    dxs.append(dx); dys.append(dy)
            dx_avg = float(np.mean(dxs)) if dxs else 0.0
            dy_avg = float(np.mean(dys)) if dys else 0.0
        else:
            dx_avg, dy_avg, _ = compute_motion_with_correlation(
                all_grids[i-1], all_grids[i])

        motion_v.append((dx_avg, dy_avg))
        xp, yp = predict_future_position(xp, yp, dx_avg, dy_avg, all_grids[i])
        pred_pos.append((xp, yp))

    return abs_pos, pred_pos, motion_v


# ─────────────────────────────────────────────────────────────────────────────
# PHOTOMETRY / SLEW
# ─────────────────────────────────────────────────────────────────────────────

def compute_photons(ext_mag: float) -> float:
    if np.isnan(ext_mag):
        return np.nan
    flux = 10 ** (-0.4 * ext_mag)
    return (PHOTON_FLUX_MAG20_ZENITH
            * RUBIN_EFFECTIVE_AREA
            * RUBIN_QUANTUM_EFFICIENCY
            * flux * EXPOSURE_TIME)


def photons_to_depth(photons: float) -> float:
    if np.isnan(photons) or photons <= 0:
        return np.nan
    snr = photons / (5 * np.sqrt(photons))
    return (20.0 - 2.5 * np.log10(snr)) if snr > 0 else np.nan


def calc_slew_time(alt1, az1, alt2: float, az2: float) -> float:
    if alt1 is None:
        return 0.0
    da  = abs(alt2 - alt1)
    daz = abs(az2  - az1)
    if daz > 180:
        daz = 360 - daz
    return max(da / MAX_SLEW_SPEED_ALT, daz / MAX_SLEW_SPEED_AZ) + SLEW_SETTLE_TIME


# ─────────────────────────────────────────────────────────────────────────────
# DREAM CSV LOADING
# ─────────────────────────────────────────────────────────────────────────────

def load_night_frames(csv_file: str, night_key: str | None = None
                       ) -> tuple[pd.DataFrame, str]:
    """
    Load cloud_sys HDF5 URLs from the DREAM CSV and filter to one night.
    If night_key is None, picks the night with the most sys frames.
    """
    df = pd.read_csv(csv_file)
    df.columns = df.columns.str.replace('"', "").str.strip()
    df = df.dropna(subset=[URL_COL]).copy()
    df[TIME_COL] = pd.to_datetime(df[TIME_COL], errors="coerce", utc=True)
    df = df.dropna(subset=[TIME_COL]).copy()
    df = df[df[URL_COL].str.contains(r"\.hdf5",   case=False, na=False)].copy()
    df = df[df[URL_COL].str.contains("cloud_sys", case=False, na=False)].copy()
    df = df.sort_values(TIME_COL).reset_index(drop=True)

    # Frames before noon UTC belong to the previous calendar night
    shifted = df[TIME_COL] - pd.Timedelta(hours=12)
    df["night_key"] = shifted.dt.date.astype(str)

    if night_key is None:
        counts    = df.groupby("night_key").size()
        night_key = str(counts.idxmax())
        print(f"  Auto-selected night: {night_key}  ({counts[night_key]} frames)")
    else:
        night_key = str(night_key)

    df_night = df[df["night_key"] == night_key].reset_index(drop=True)
    print(f"  Night {night_key}: {len(df_night)} cloud_sys frames available")
    return df_night, night_key


def load_dream_frames(df_night: pd.DataFrame, max_frames: int | None = None
                       ) -> tuple[list, list, list, np.ndarray, np.ndarray]:
    """
    Fetch every DREAM HDF5 frame and project onto the flat-sky km grid.
    Returns (ext_grids, sig_grids, metas, x_edges, y_edges).
    """
    ext_grids, sig_grids, metas = [], [], []
    x_edges = y_edges = None
    rows = df_night.iloc[:max_frames] if max_frames else df_night

    for i, (_, row) in enumerate(rows.iterrows()):
        url = transform_url(row[URL_COL])
        try:
            clouds, sigma = fetch_sys_map(url)
            t = row[TIME_COL].to_pydatetime()
            ext_g, sig_g, _, _, xe, ye = process_to_grid(clouds, sigma, t)
            ext_grids.append(ext_g)
            sig_grids.append(sig_g)
            metas.append({"time": t, "mjd": Time(t).mjd})
            if x_edges is None:
                x_edges, y_edges = xe, ye
            if (i + 1) % 50 == 0:
                print(f"    Loaded {i+1} / {len(rows)} frames …")
        except Exception as e:
            print(f"    Frame {i} skipped: {e}")

    print(f"  {len(ext_grids)} frames loaded successfully")
    return ext_grids, sig_grids, metas, x_edges, y_edges


# ─────────────────────────────────────────────────────────────────────────────
# RUBIN SCHEDULER SIMULATION
# ─────────────────────────────────────────────────────────────────────────────

def standard_survey_footprint(survey_start_mjd: float, nside: int = NSIDE):
    sky = CurrentAreaMap(nside=nside)
    footprints_hp_array, labels = sky.return_maps()
    roll_indx = np.where((labels == "lowdust") | (labels == "virgo"))[0]
    footprints_hp = {key: footprints_hp_array[key]
                     for key in footprints_hp_array.dtype.names}
    almanac  = Almanac(mjd_start=survey_start_mjd)
    sun_info = almanac.get_sun_moon_positions(survey_start_mjd)
    sun_ra   = sun_info["sun_RA"].copy()
    return make_rolling_footprints(
        fp_hp=footprints_hp, mjd_start=survey_start_mjd,
        sun_ra_start=sun_ra, nslice=2, scale=0.9, nside=nside,
        wfd_indx=roll_indx, order_roll=1, n_cycles=3, uniform=True)


def build_greedy_surveys(footprints, nside: int = NSIDE, band: str = "r"):
    bfs = [
        (bf.M5DiffBasisFunction(bandname=band, nside=nside), 3.0),
        (bf.FootprintBasisFunction(bandname=band, footprint=footprints,
                                    out_of_bounds_val=np.nan, nside=nside), 0.75),
        (bf.SlewtimeBasisFunction(bandname=band, nside=nside), 3.0),
        (bf.BandChangeBasisFunction(bandname=band), 100.0),
        (bf.BandLoadedBasisFunction(bandnames=[band]), 0),
        (bf.VisitRepeatBasisFunction(gap_min=0, gap_max=120.0,
                                      bandname=None, nside=nside, npairs=20), -1.0),
        (bf.AltAzShadowMaskBasisFunction(nside=nside, shadow_minutes=15.0,
                                          max_alt=76.0, pad=3.0), 0),
        (bf.MoonAvoidanceBasisFunction(nside=nside, moon_distance=30), 0),
        (bf.PlanetMaskBasisFunction(nside=nside), 0),
    ]
    weights      = [v[1] for v in bfs]
    basis        = [v[0] for v in bfs]
    detailer_lst = [
        detailers.CameraRotDetailer(min_rot=min(CAMERA_ROT_LIMITS),
                                     max_rot=max(CAMERA_ROT_LIMITS)),
        detailers.LabelRegionsAndDDFs(),
    ]
    return [GreedySurvey(
        basis, weights,
        exptime=EXPOSURE_TIME, bandname=band, nside=nside, nexp=1,
        detailers=detailer_lst,
        survey_name=f"greedy_{band}",
        science_program=SCIENCE_PROGRAM,
        observation_reason=f"singles_{band}",
        ignore_obs=["DD", "twilight_near_sun", "ToO"],
        block_size=1, smoothing_kernel=None,
        seed=42, camera="LSST", dither="night",
    )]


def run_scheduler_simulation(survey_start_mjd: float,
                              sim_duration_days: float = 1.0,
                              verbose: bool = True) -> np.ndarray:
    print(f"\n{'='*60}")
    print("RUNNING GREEDY SCHEDULER SIMULATION (ideal clouds)")
    print(f"{'='*60}")
    footprint  = standard_survey_footprint(survey_start_mjd)
    surveys    = build_greedy_surveys(footprint)
    cloud_maps = CloudMap()
    cloud_maps.add_frame(np.zeros(hp.nside2npix(NSIDE)), survey_start_mjd)
    scheduler  = CoreScheduler(surveys, nside=NSIDE)
    mo = ModelObservatory(
        nside=NSIDE, mjd_start=survey_start_mjd,
        cloud_data="ideal", downtimes="ideal",
        cloud_maps=cloud_maps)
    mo, scheduler, obs = sim_runner(
        mo, scheduler, sim_duration=sim_duration_days, verbose=verbose)
    print(f"  Scheduler produced {len(obs)} observations")
    return obs


# ─────────────────────────────────────────────────────────────────────────────
# OBSERVATION FIELD ACCESSORS
# Rubin sim_runner returns a numpy structured array; field names vary by version.
# ─────────────────────────────────────────────────────────────────────────────

def _obs_mjd(obs_row) -> float:
    """Extract MJD — handles multiple rubin_scheduler field name conventions."""
    for field in ("observationStartMJD", "start_mjd", "mjd",
                  "MJD", "observationStartTime"):
        try:
            return float(obs_row[field])
        except (KeyError, ValueError, IndexError):
            pass
    raise KeyError(
        f"No MJD field found. Available fields: {obs_row.dtype.names}")


def _obs_radec(obs_row) -> tuple[float, float]:
    """
    Extract (RA_deg, Dec_deg).
    rubin_scheduler stores RA/Dec in RADIANS as 'RA' and 'dec'.
    """
    for ra_f, dec_f in [("RA", "dec"), ("fieldRA", "fieldDec"), ("ra", "dec")]:
        try:
            ra  = float(obs_row[ra_f])
            dec = float(obs_row[dec_f])
            # If values are within radian range, convert to degrees
            if -2*np.pi - 0.1 <= ra <= 2*np.pi + 0.1:
                ra  = np.degrees(ra) % 360.0
                dec = np.degrees(dec)
            return ra, dec
        except (KeyError, ValueError, IndexError):
            pass
    raise KeyError(
        f"No RA/Dec fields found. Available: {obs_row.dtype.names}")


# ─────────────────────────────────────────────────────────────────────────────
# METRICS COMPUTATION
# ─────────────────────────────────────────────────────────────────────────────

def _empty_m() -> dict:
    return {"photons": [], "expose_times": [], "slew_times": [],
            "extinctions": [], "depths": [], "frame_indices": [],
            "visited_pix": set()}


def _record(m: dict, photons_full: float, ext: float, sl: float,
            fi: int, alt: float, az: float,
            prev_alt: list, prev_az: list) -> None:
    """Append one observation slot; slew-gate the photon count."""
    expose   = max(0.0, SLOT_TIME - sl)
    ph_gated = photons_full * (expose / EXPOSURE_TIME)
    m["photons"].append(ph_gated)
    m["expose_times"].append(expose)
    m["slew_times"].append(sl)
    m["extinctions"].append(ext)
    m["depths"].append(photons_to_depth(ph_gated))
    m["frame_indices"].append(fi)
    pix = hp.ang2pix(NSIDE, np.radians(90.0 - alt), np.radians(az), nest=False)
    m["visited_pix"].add(pix)
    prev_alt[0] = alt
    prev_az[0]  = az


def compute_dream_metrics(ext_grids: list[np.ndarray],
                           sig_grids: list[np.ndarray],
                           abs_positions: list,
                           pred_positions: list) -> dict[str, dict]:
    """
    Per-frame metrics for both DREAM strategies.
    abs_positions / pred_positions come from compute_all_positions().
    """
    print(f"\n{'='*60}")
    print("COMPUTING DREAM STRATEGY METRICS")
    print(f"{'='*60}")

    strats = {k: _empty_m() for k in ("absolute", "motion")}
    pa, paa = [None], [None]   # mutable prev_alt / prev_az for absolute
    pm, pma = [None], [None]   # same for motion

    for fi, (grid, sgrid) in enumerate(zip(ext_grids, sig_grids)):

        # ── Absolute minimum ─────────────────────────────────────────────────
        xa, ya   = abs_positions[fi]
        alt_a, az_a = xy_to_altaz(xa, ya)
        ext_a    = get_value_at_position(grid, xa, ya)
        if alt_a >= MIN_ALT_DEG and not np.isnan(ext_a):
            sl = calc_slew_time(pa[0], paa[0], alt_a, az_a)
            _record(strats["absolute"], compute_photons(ext_a),
                    ext_a, sl, fi, alt_a, az_a, pa, paa)

        # ── Motion tracking ──────────────────────────────────────────────────
        xp, yp   = pred_positions[fi]
        alt_p, az_p = xy_to_altaz(xp, yp)
        ext_p    = get_value_at_position(grid, xp, yp)
        if alt_p >= MIN_ALT_DEG and not np.isnan(ext_p):
            sl = calc_slew_time(pm[0], pma[0], alt_p, az_p)
            _record(strats["motion"], compute_photons(ext_p),
                    ext_p, sl, fi, alt_p, az_p, pm, pma)

        if (fi + 1) % 50 == 0:
            print(f"  Frame {fi+1}/{len(ext_grids)}")

    for k, m in strats.items():
        print(f"  {k}: {len(m['photons'])} valid frames, "
              f"{len(m['visited_pix'])} unique HEALPix pixels")
    return strats


def compute_scheduler_metrics(observations: np.ndarray,
                               ext_grids: list[np.ndarray],
                               sig_grids: list[np.ndarray],
                               metas: list[dict]) -> dict:
    """
    Map real scheduler pointings onto DREAM cloud grids.
    Each observation is matched to the nearest cloud frame by MJD.
    RA/Dec converted to Alt/Az via proper astropy transform at that timestamp.
    """
    print(f"\n{'='*60}")
    print("COMPUTING SCHEDULER METRICS AGAINST REAL DREAM CLOUDS")
    print(f"{'='*60}")
    print(f"  obs dtype fields: {observations.dtype.names}")

    mjd_arr   = np.array([m["mjd"] for m in metas])
    m_sched   = _empty_m()
    pa, paa   = [None], [None]
    n_below, n_nomap = 0, 0

    for obs in observations:
        # ── Extract MJD and sky coords ───────────────────────────────────────
        try:
            obs_mjd       = _obs_mjd(obs)
            ra_deg, dec_deg = _obs_radec(obs)
        except KeyError as e:
            print(f"  FATAL schema error: {e}")
            break

        # ── Match to nearest cloud frame ─────────────────────────────────────
        fi    = int(np.argmin(np.abs(mjd_arr - obs_mjd)))
        t     = metas[fi]["time"]
        grid  = ext_grids[fi]
        sgrid = sig_grids[fi]

        # ── Proper coordinate transform ──────────────────────────────────────
        alt_deg, az_deg = radec_to_altaz(ra_deg, dec_deg, t)
        x, y = altaz_to_xy(alt_deg, az_deg)
        if x is None:
            n_below += 1
            continue

        # ── Cloud extinction at this pointing ────────────────────────────────
        ext = get_value_at_position(grid, x, y)
        if np.isnan(ext):
            # Scheduler pointed outside cloud map coverage → assume photometric
            ext = 0.0
            n_nomap += 1

        sl = calc_slew_time(pa[0], paa[0], alt_deg, az_deg)
        _record(m_sched, compute_photons(ext),
                ext, sl, fi, alt_deg, az_deg, pa, paa)

    print(f"  Processed: {len(m_sched['photons'])} observations")
    print(f"  Rejected (below alt {MIN_ALT_DEG}°): {n_below}")
    print(f"  Outside cloud map (assumed photometric): {n_nomap}")
    return m_sched


# ─────────────────────────────────────────────────────────────────────────────
# SUMMARY STATISTICS
# ─────────────────────────────────────────────────────────────────────────────

def summarize(label: str, m: dict) -> dict:
    ph    = np.array(m["photons"])
    et    = np.array(m["expose_times"])
    sl    = np.array(m["slew_times"])
    ex    = np.array(m["extinctions"])
    dp    = np.array([d for d in m["depths"] if not np.isnan(d)])
    slots = len(ph) * SLOT_TIME

    row = {
        "Strategy":            label,
        "N_slots":             len(ph),
        "Total_photons":       float(np.sum(ph)),
        "Shutter_eff_%":       float(np.sum(et) / slots * 100) if slots > 0 else 0.0,
        "Mean_slew_s":         float(np.mean(sl[1:])) if len(sl) > 1 else 0.0,
        "Mean_extinction_mag": float(np.nanmean(ex)),
        "Median_depth_mag":    float(np.median(dp)) if len(dp) else np.nan,
        "Unique_pixels":       len(m["visited_pix"]),
    }

    print(f"\n  ── {label} ──")
    for k, v in row.items():
        if k == "Strategy":
            continue
        print(f"    {k:<28} {v:.5g}")
    return row


# ─────────────────────────────────────────────────────────────────────────────
# PLOTS
# ─────────────────────────────────────────────────────────────────────────────

def plot_comparison(all_metrics: dict[str, dict],
                    summary_rows: list[dict],
                    ext_grids: list[np.ndarray],
                    x_edges: np.ndarray,
                    y_edges: np.ndarray,
                    night_tag: str,
                    output_dir: str = "outputs"):
    os.makedirs(output_dir, exist_ok=True)
    print(f"\n{'='*60}")
    print("CREATING COMPARISON PLOTS")
    print(f"{'='*60}")

    labels = list(all_metrics.keys())
    colors = {
        "DREAM Absolute": "#2ca02c",
        "DREAM Motion":   "#1f77b4",
        "Scheduler":      "#d62728",
    }

    # ── Figure 1: Main comparison panel ─────────────────────────────────────
    fig = plt.figure(figsize=(21, 14))
    gs  = fig.add_gridspec(3, 3, hspace=0.42, wspace=0.32)

    def _bar(ax, key: str, ylabel: str, invert: bool = False,
             pct_vs: int | None = None, fmt: str = ".3g"):
        vals = [r[key] for r in summary_rows]
        bars = ax.bar(labels, vals,
                      color=[colors[l] for l in labels],
                      alpha=0.82, edgecolor="black", lw=1.2)
        for i, (bar, v) in enumerate(zip(bars, vals)):
            lbl_txt = format(v, fmt)
            if pct_vs is not None and i != pct_vs:
                base    = vals[pct_vs]
                delta   = (v - base) / max(abs(base), 1) * 100
                lbl_txt += f"\n({delta:+.1f}%)"
            ypos = v * 1.01 if not invert else v - abs(v) * 0.03
            ax.text(bar.get_x() + bar.get_width() / 2, ypos,
                    lbl_txt, ha="center", va="bottom", fontsize=9, weight="bold")
        ax.set_ylabel(ylabel)
        ax.grid(alpha=0.3, axis="y")
        if invert:
            ax.invert_yaxis()

    ax = fig.add_subplot(gs[0, 0])
    _bar(ax, "Total_photons", "Total photons (slew-gated)", pct_vs=2)
    ax.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))
    ax.set_title("Total Night Photon Collection", weight="bold")

    ax = fig.add_subplot(gs[0, 1])
    _bar(ax, "Shutter_eff_%", "Shutter-open efficiency (%)", fmt=".1f")
    ax.set_ylim(0, 110)
    ax.set_title("Observation Efficiency", weight="bold")

    ax = fig.add_subplot(gs[0, 2])
    _bar(ax, "Median_depth_mag", "Median 5σ depth (mag)", invert=True, fmt=".2f")
    ax.set_title("Survey Depth (slew-scaled)", weight="bold")

    # Photon timeline
    ax = fig.add_subplot(gs[1, :])
    for lbl, c in colors.items():
        fi = all_metrics[lbl]["frame_indices"]
        ph = all_metrics[lbl]["photons"]
        if fi:
            ax.plot(fi, ph, color=c, alpha=0.8, lw=1.5, label=lbl)
    ax.set_xlabel("Cloud frame index (time →)")
    ax.set_ylabel("Slew-gated photons per 32-s slot")
    ax.set_title(f"Photon Collection Over the Night — {night_tag} "
                 "(real DREAM cloud_sys data)", weight="bold")
    ax.legend(fontsize=10); ax.grid(alpha=0.3)
    ax.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))

    ax = fig.add_subplot(gs[2, 0])
    for lbl, c in colors.items():
        sl = all_metrics[lbl]["slew_times"][1:]
        if sl:
            ax.hist(sl, bins=30, alpha=0.6, color=c, label=lbl,
                    edgecolor="black", lw=0.5)
    ax.set_xlabel("Slew time (s)"); ax.set_ylabel("Count")
    ax.set_title("Slew Time Distribution", weight="bold")
    ax.legend(fontsize=9); ax.grid(alpha=0.3)

    ax = fig.add_subplot(gs[2, 1])
    for lbl, c in colors.items():
        ex = all_metrics[lbl]["extinctions"]
        if ex:
            ax.hist(ex, bins=30, alpha=0.6, color=c, label=lbl,
                    edgecolor="black", lw=0.5)
    ax.set_xlabel("Extinction experienced (mag)"); ax.set_ylabel("Count")
    ax.set_title("Extinction Distribution per Strategy", weight="bold")
    ax.legend(fontsize=9); ax.grid(alpha=0.3)

    ax = fig.add_subplot(gs[2, 2])
    _bar(ax, "Unique_pixels", "Unique HEALPix pixels visited")
    ax.set_title("Sky Coverage", weight="bold")

    plt.suptitle(
        f"Rubin Observatory Pointing Strategy Comparison — Night {night_tag}\n"
        "DREAM Absolute Min  ·  DREAM Motion Tracking  ·  Greedy Scheduler\n"
        "(real DREAM cloud_sys HDF5 data · σ-masked pixels · slew-gated photons)",
        fontsize=13, weight="bold", y=1.005)
    out1 = os.path.join(output_dir, f"comparison_metrics_{night_tag}.png")
    plt.savefig(out1, dpi=150, bbox_inches="tight")
    plt.close()
    print(f"  Saved: {out1}")

    # ── Figure 2: Cloud map snapshots ────────────────────────────────────────
    n   = len(ext_grids)
    fis = [0, n // 4, n // 2, 3 * n // 4, n - 1]
    fig, axes = plt.subplots(1, len(fis), figsize=(5 * len(fis), 5))
    for ax, fi, title in zip(axes, fis, ["Start", "25%", "Mid", "75%", "End"]):
        im  = ax.pcolormesh(x_edges, y_edges, ext_grids[fi],
                            cmap="viridis", vmin=-0.2, vmax=2.0, shading="flat")
        plt.colorbar(im, ax=ax, label="Extinction (mag)")
        th  = np.linspace(0, 2*np.pi, 300)
        ax.plot(15_000*np.cos(th), 15_000*np.sin(th), "w-", lw=1, alpha=0.4)
        ax.plot(0, 0, "r+", ms=12, mew=2.5)
        ax.set_title(f"{title}  (frame {fi})", weight="bold")
        ax.set_aspect("equal")
        ax.set_xlabel("X (km)"); ax.set_ylabel("Y (km)")
    plt.suptitle(f"Real DREAM Cloud Maps — Night {night_tag}", fontsize=13, weight="bold")
    out2 = os.path.join(output_dir, f"cloud_maps_{night_tag}.png")
    plt.savefig(out2, dpi=150, bbox_inches="tight")
    plt.close()
    print(f"  Saved: {out2}")

    # ── Figure 3: Mollweide coverage maps ────────────────────────────────────
    npix_hp = hp.nside2npix(NSIDE)
    fig = plt.figure(figsize=(21, 6))
    for idx, (lbl, c) in enumerate(colors.items()):
        cov = np.zeros(npix_hp)
        for p in all_metrics[lbl]["visited_pix"]:
            cov[p] = 1.0
        hp.mollview(cov, fig=fig.number, sub=(1, 3, idx + 1),
                    title=f"{lbl}\n({len(all_metrics[lbl]['visited_pix'])} pixels)",
                    cmap="YlOrRd", min=0, max=1, hold=True)
    plt.suptitle(f"Sky Coverage Maps — Night {night_tag}",
                 fontsize=13, weight="bold")
    out3 = os.path.join(output_dir, f"coverage_map_{night_tag}.png")
    plt.savefig(out3, dpi=150, bbox_inches="tight")
    plt.close()
    print(f"  Saved: {out3}")

    return out1, out2, out3


# ─────────────────────────────────────────────────────────────────────────────
# MAIN ENTRY POINT
# ─────────────────────────────────────────────────────────────────────────────

def main(
    csv_file:          str        = "feb5_data.csv",
    night_key:         str | None = None,
    sim_duration_days: float      = 1.0,
    max_frames:        int | None = None,
    output_dir:        str        = "outputs",
    verbose:           bool       = True,
):
    """
    Full pipeline using REAL DREAM cloud data:

      1. Load cloud_sys HDF5 frames from DREAM CSV for the target night
      2. Grid every HEALPix map → flat-sky extinction + sigma grids
      3. Run cloud motion tracking (compute_all_positions)
      4. Run rubin_scheduler greedy simulation (ideal clouds, same night MJD)
      5. Compute per-slot metrics for all three strategies against REAL clouds
      6. Print summary table and save plots

    Parameters
    ----------
    csv_file          : Path to the DREAM CSV (e.g. 'feb5_data.csv')
    night_key         : 'YYYY-MM-DD' to pick a specific night;
                        None = auto-select night with most sys frames
    sim_duration_days : How long to simulate the scheduler (days)
    max_frames        : Cap frames loaded (useful for quick testing)
    output_dir        : Where to write plots and CSV
    verbose           : Pass to sim_runner
    """
    os.makedirs(output_dir, exist_ok=True)

    # ── Step 1: Index CSV and pick night ─────────────────────────────────────
    print(f"\n{'='*60}")
    print("LOADING REAL DREAM CLOUD DATA")
    print(f"{'='*60}")
    df_night, night_key = load_night_frames(csv_file, night_key)
    if len(df_night) < 5:
        raise RuntimeError(
            f"Only {len(df_night)} frames for night {night_key} — aborting.")
    night_tag = str(night_key).replace("-", "")

    # ── Step 2: Load + grid all frames ───────────────────────────────────────
    print(f"\n{'='*60}")
    print("GRIDDING HEALPIX → FLAT-SKY (one astropy transform per frame)")
    print(f"{'='*60}")
    ext_grids, sig_grids, metas, x_edges, y_edges = load_dream_frames(
        df_night, max_frames=max_frames)
    if len(ext_grids) < 2:
        raise RuntimeError("Too few frames loaded — check S3 credentials / network access.")

    # ── Step 3: Cloud motion tracking ────────────────────────────────────────
    print(f"\n{'='*60}")
    print(f"CLOUD MOTION TRACKING — {len(ext_grids)} frames")
    print(f"{'='*60}")
    abs_pos, pred_pos, _ = compute_all_positions(ext_grids)

    # ── Step 4: Scheduler simulation ─────────────────────────────────────────
    # Use noon UTC before the first cloud frame as survey start
    survey_mjd = metas[0]["mjd"] - 0.5
    sched_obs  = run_scheduler_simulation(
        survey_start_mjd  = survey_mjd,
        sim_duration_days = sim_duration_days,
        verbose           = verbose)

    # ── Step 5: Metrics ──────────────────────────────────────────────────────
    dream_m = compute_dream_metrics(ext_grids, sig_grids, abs_pos, pred_pos)
    sched_m = compute_scheduler_metrics(sched_obs, ext_grids, sig_grids, metas)

    all_metrics = {
        "DREAM Absolute": dream_m["absolute"],
        "DREAM Motion":   dream_m["motion"],
        "Scheduler":      sched_m,
    }

    # ── Step 6: Print summary ────────────────────────────────────────────────
    print(f"\n{'='*60}")
    print("NIGHT PERFORMANCE SUMMARY")
    print(f"{'='*60}")
    summary_rows = [summarize(lbl, m) for lbl, m in all_metrics.items()]

    baseline = summary_rows[2]["Total_photons"]
    print(f"\n  Δ photons vs Scheduler baseline:")
    for r in summary_rows[:2]:
        pct = (r["Total_photons"] - baseline) / max(abs(baseline), 1) * 100
        print(f"    {r['Strategy']:<22}  {pct:+.1f}%")

    csv_out = os.path.join(output_dir, f"summary_table_{night_tag}.csv")
    pd.DataFrame(summary_rows).to_csv(csv_out, index=False)
    print(f"\n  Summary CSV → {csv_out}")

    # ── Step 7: Plots ────────────────────────────────────────────────────────
    plot_comparison(all_metrics, summary_rows,
                    ext_grids, x_edges, y_edges,
                    night_tag, output_dir)

    print(f"\n{'='*60}")
    print(f"DONE — all outputs in: {output_dir}/")
    print(f"{'='*60}")
    return all_metrics, summary_rows


# ─────────────────────────────────────────────────────────────────────────────
# CLI
# ─────────────────────────────────────────────────────────────────────────────

if __name__ == "__main__":
    import sys

    # Detect Jupyter / IPython kernel — argparse cannot run there
    _in_jupyter = (
        "ipykernel" in sys.modules
        or any("jupyter" in a or "ipykernel" in a for a in sys.argv)
    )

    if _in_jupyter:
        # ── Running inside a Jupyter notebook cell ───────────────────────────
        # Edit these defaults directly and call main() below.
        main(
            csv_file          = "feb5_data.csv",   # ← your CSV here
            night_key         = None,               # None = auto (most frames)
            sim_duration_days = 1.0,
            max_frames        = None,               # set e.g. 30 for a fast test
            output_dir        = "outputs",
            verbose           = True,
        )
    else:
        # ── Running as a plain script: python dream_scheduler_comparison.py ──
        ap = argparse.ArgumentParser(
            description="Compare DREAM cloud-aware strategies vs the greedy scheduler")
        ap.add_argument("--csv",        default="feb5_data.csv")
        ap.add_argument("--night",      default=None)
        ap.add_argument("--outdir",     default="outputs")
        ap.add_argument("--sim-days",   type=float, default=1.0)
        ap.add_argument("--max-frames", type=int,   default=None)
        ap.add_argument("--quiet",      action="store_true")
        args = ap.parse_args()

        main(
            csv_file          = args.csv,
            night_key         = args.night,
            sim_duration_days = args.sim_days,
            max_frames        = args.max_frames,
            output_dir        = args.outdir,
            verbose           = not args.quiet,
        )
```
